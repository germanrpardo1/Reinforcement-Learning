## Q1: Calculating the Value Function (5%)
Consider a larger version of the Random Walk problem we discussed in Seminar 4. 
Now we have n non-terminal states instead of 5, i.e., 

F -- 1 -- 2 -- 3 -- 4 -- 5 -- ... -- n-1 -- n -- S.

Arriving at the left terminal state (F) gives a reward of -1 whereas reaching the right terminal state (S) gives a reward of 1. All other transitions give a reward of 0. 

**Task**: Calculate the value function for each non-terminal state and provide a justification for your answers. For this question, coding is not required. Please provide a closed-form expression for the value functions, alongside your mathematical calculations to illustrate your process.

### First case to obtain the general pattern
Consider the same example of the seminar for $n=3$ non-terminal states (with $\gamma = 0$) and reward of 1 for reaching state (S) and 0 for reaching any other state (including F):


$$
\begin{aligned}
V^\pi(1) &= \frac{1}{2} V^\pi(2)\\
V^\pi(2) &= \frac{1}{2} V^\pi(1) + \frac{1}{2} V^\pi(3)\\
V^\pi(3) &= \frac{1}{2} + \frac{1}{2} V^\pi(2)
\end{aligned}
$$

We can solve this the following way, by replacing:

$$
\begin{aligned}
V^\pi(2) &= \frac{1}{2} \left[\frac{1}{2} V^\pi(2)\right] + \frac{1}{2} \left[\frac{1}{2} + \frac{1}{2} V^\pi(2)\right]\\
V^\pi(2) &= \frac{1}{4} V^\pi(2) + \frac{1}{4} + \frac{1}{4} V^\pi(2)\\
\frac{1}{2}V^\pi(2) &= \frac{1}{4}
\end{aligned}
$$

Which yields to the following solution:

$$
\begin{aligned}
&V^\pi(F) = 0\\
&V^\pi(1) = \frac{1}{4}\\
&V^\pi(2) = \frac{2}{4} = \frac{1}{2}\\
&V^\pi(3) = \frac{3}{4}\\
&V^\pi(S) = 1
\end{aligned}
$$

By looking at this, one can notice that it is possible to generalise the value function for any $i | 0\leq i \leq n+1$. Let $N$ be the total number of states (for this small case $N=5$) and let $n$ be the number of the non-terminal states (i.e., $ n = 3$). Therefore, the value function for any state can be formulated as:

$$
V^\pi(i) = \frac{i}{N-1}, \quad  \forall i = 0, ..., n+1
$$

### Second case (actual question)
Having this simple case in mind, now consider the same problem but the reward for reaching state F is now $-1$, the reward for reaching state S is $1$ and for the rest the reward is 0 (just as asked in this question). Now, the value function for $n=3$ can be written similarly as:

$$
\begin{aligned}
V^\pi(1) &= -\frac{1}{2} + \frac{1}{2} V^\pi(2)\\
V^\pi(2) &= \frac{1}{2} V^\pi(1) + \frac{1}{2} V^\pi(3)\\
V^\pi(3) &= \frac{1}{2} + \frac{1}{2} V^\pi(2)
\end{aligned}
$$

Now I replace in the same way as before:

$$
\begin{aligned}
V^\pi(2) = \frac{1}{2} \left[-\frac{1}{2} + \frac{1}{2} V^\pi(2)\right] + \frac{1}{2} \left[\frac{1}{2} + \frac{1}{2} V^\pi(2)\right]\\
V^\pi(2) = -\frac{1}{4} + \frac{1}{4} V^\pi(2) + \frac{1}{4} + \frac{1}{4} V^\pi(2)
\end{aligned}
$$

Which yields to:

$$
\begin{aligned}
&V^\pi(F) = -1\\
&V^\pi(1) = -\frac{1}{2}\\
&V^\pi(2) = 0\\
&V^\pi(3) = \frac{1}{2}\\
&V^\pi(S) = 1
\end{aligned}
$$

The important thing to notice here is that (as in the first case) the value function is linear with respect to its states, so it is just a matter of finding the linear function to determine their value in a general way.
Having said this, consider now the relationship with the first and second case, one can say that the value functions from the second case are a linear transformation of the value functions of the second case, therefore:

For $V^\pi(n)$, the value function times a slope $m$ and adding a constant $b$:

$$
\begin{aligned}
&V^\pi(1): \frac{1}{4} m + b = -\frac{1}{2}\\
&V^\pi(2): \frac{1}{2} m + b = 0\\
&V^\pi(3): \frac{3}{2} m + b = \frac{1}{2}
\end{aligned}
$$

By solving this set of equations, one obtain the following:

$$
m = 2\\
b = -1
$$

Now, I bring back the equation for calculating the value function of the first case:

$$
V^\pi(i) = \frac{i}{N-1}, \quad  \forall i = 0,..., n+1
$$

Notice that if I multiply this equation by $m=2$ and substract the constant $b=-1$, I will obtain the value function for the second case for any $i\in \{0, ..., n+1\}$. This way:

$$
V^\pi(i) = \frac{2}{N-1}i-1, \quad  \forall i = 0,..., n+1\\
\vdots\\
V^\pi(i) =\frac{2i-N+1}{N-1}, \quad  \forall i = 0,..., n+1
$$

### Final answer 

Therefore, the final answer to this question is the following expression that calculates the value function in a 1 dimensional random walk with reward -1 in one terminal state, reward 1 in the other terminal state and 0 reward for the other states.


$$
V^\pi(i) =\frac{2i-N+1}{N-1}, \quad  \forall i = 0,..., n+1
$$

(Recall that $N$ stands for the $total$ number of states (i.e., $n+2$), and $n$ is the number of $non$-$terminal$ states).
